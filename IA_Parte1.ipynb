```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Percorso del file TIFF
tiff_path = "/content/13N-105W-40N-L3H-SR-2018_01_01.tif"



# Apertura del file TIFF
with rasterio.open(tiff_path) as dataset:
    # Leggi le bande del dataset Sentinel-2
    red_band = dataset.read(3)  # Banda rossa
    green_band = dataset.read(2)  # Banda verde
    blue_band = dataset.read(1)  # Banda blu

# Normalizzazione delle bande
red_band = (red_band - np.min(red_band)) / (np.max(red_band) - np.min(red_band))
green_band = (green_band - np.min(green_band)) / (np.max(green_band) - np.min(green_band))
blue_band = (blue_band - np.min(blue_band)) / (np.max(blue_band) - np.min(blue_band))

# Combinazione delle bande in un'immagine RGB
rgb_image = np.dstack((red_band, green_band, blue_band))


# Creazione di un oggetto immagine PIL
pil_image = Image.fromarray((rgb_image * 255).astype(np.uint8))

# Salvataggio dell'immagine come file PNG
png_path = "percorso_del_tuo_file.png"
pil_image.save(png_path)

# Visualizzazione dell'immagine
plt.imshow(pil_image)
plt.show()
```
